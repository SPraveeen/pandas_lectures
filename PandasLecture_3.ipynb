```python
## Q- find the details of the movies by high budget directors.
# "High budget directors" -> atleast 1 movie with budget >=100M
```

```python
data_dir_budget = data.groupby('director_name')['budget'].max().reset_index()
data_dir_budget.head()
```

```python
names = data_dir_budget[data_dir_budget['budget']>= 100000000]["director_name"]
names
```

```python
data[data['director_name'].isin(names)]
```

```python
# filter
```

```python
def isHighBudget(dir):
    return dir['budget'].max() >= 100000000
```

```python
data.groupby('director_name').filter(isHighBudget)
```

```python
# Risky movie ?
# - budget of the movie > the avg budget of the director
```

```python
# step 1 : avg budget of the director.
# step 2: budget of the movie  > avg budget of director
# step 3: create a new col data['risky]
```

```python
def risky(x):
    x['risky'] =  x['budget'] - x['budget'].mean() >= 0
    return x

data = data.groupby('director_name').apply(risky)
```
