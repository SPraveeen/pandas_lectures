```python
A = [[1,2,3],
     [4,5,6],
     [7,8,9]]

```

```python
nr = len(A)
nc = len(A[0])
```

```python
total = 0
for i in range(nr):
    for j in range(nc):
        if i == j:
            total += A[i][j]


print(total)
```
