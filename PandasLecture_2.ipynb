```python
users = pd.DataFrame({"userid":[1, 2, 3], "name":["sharadh", "shahid", "khusalli"]})
users
```

```python
msgs = pd.DataFrame({"userid":[1, 1, 2, 4], "msg":['hmm', "acha", "theek hai", "nice"]})
msgs
```

```python
pd.concat([users, msgs], ignore_index=True)
```

```python
pd.concat([users, msgs], axis=1)
```

```python
users
```

```python
msgs
```

```python
users.merge(msgs, how='inner')
```

```python
# check users data and their msgs if available.

users.merge(msgs, how='left')
```

```python
# all msgs and theirs user if available.

users.merge(msgs, how='right')
```

```python
users.merge(msgs, how='outer')
```

```python
users.merge(msgs, how='outer', on='userid')
```

```python
users.rename({'userid' : 'id'}, inplace = True, axis = 1)
users
```

```python
msgs
```

```python
users.merge(msgs, how='outer', on='userid')
```

```python
users.merge(msgs, how='inner', left_on='id', right_on='userid')
```
